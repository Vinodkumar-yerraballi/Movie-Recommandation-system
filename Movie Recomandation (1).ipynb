```python
def make_genresList(x):
    gen = []
    st = " "
    for i in x:
        if i.get('name') == 'Science Fiction':
            scifi = 'Sci-Fi'
            gen.append(scifi)
        else:
            gen.append(i.get('name'))
    if gen == []:
        return np.NaN
    else:
        return (st.join(gen))
```

```python
data['spoken_languages']=data['spoken_languages'].map(lambda x:make_genresList(x))
data['spoken_languages']
```

```python
data['genres'] = data['genres'].replace(np.nan, 'unknown')
data['spoken_languages'] = data['spoken_languages'].replace(np.nan, 'unknown')
data['overview'] = data['overview'].replace(np.nan, 'unknown')
data['tagline'] = data['tagline'].replace(np.nan, 'unknown')
data['runtime'] = data['runtime'].replace(np.nan, 'unknown')
data['cast'] = data['cast'].replace(np.nan, 'unknown')
data['director'] = data['director'].replace(np.nan, 'unknown')
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
def create_soup(data):
    # Creating a simple text for countvectorizer to work with
    att = data['title'].lower()
    for i in data[1:]:
        att = att + ' ' + str(i)
    return att
 
model_data=data.copy()
model_data=model_data[['title','keywords','id','genres','cast','director']]
model_data['soup']=model_data.apply(create_soup,axis=1)
```

```python
cv=CountVectorizer(stop_words='english')
vector=cv.fit_transform(model_data['soup']).toarray()
```

```python
vector
```

```python
cv.get_feature_names()
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stemmer(x):
    stem_word=[]
    string=" "
    for i in x.split():
        stem_word.append(ps.stem(i))
    return string.join(stem_word)
```

```python
model_data['soup']=model_data['soup'].map(lambda x:stemmer(x))
```

```python
model_data.head()
```

```python
model_data.to_csv('final_data.csv')
```

```python
vectors=cv.fit_transform(model_data['soup']).toarray()
```

```python
vectors
```

```python
cv.get_feature_names()
```

```python
similarity=cosine_similarity(vectors)
```

```python
similarity[1]
```

```python
def recommend(movie):
    movie_index = model_data[model_data['title'] == movie].index[0]
    distances = similarity[movie_index]
    recommended_movies_list = sorted(list(enumerate(distances)),reverse = True,key = lambda x: x[1])[1:9]
    
    for i in recommended_movies_list:
        print(model_data.iloc[i[0]].title)
```

```python
recommend('The Dark Knight Rises')
```
